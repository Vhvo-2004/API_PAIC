## Manejo dos dados para o banco

Existe a possibilidade de que essas informações que não possuem campo no banco atualmente possam ser relevantes em algum momento durante o desenvolver da pesquisa ,por isso optou-se por construir um banco responsável por manter as informações no formato original e um banco com o formato otimizado para o que se precisa para a aplicação.(O nome dessa estrategia é data Warehouse).



##### Criando o novo banco de dados

```python
# Criando o banco com os dados do CSV (ignorando linhas que estao em formato invalido)
import pandas as pd
from sqlalchemy import create_engine
from pathlib import Path


```

```python
# ======= CONFIGURE AQUI =======
MYSQL_DSN = "mysql+pymysql://root:@127.0.0.1:3306/raw" # ajuste
CSV_MAP = {
    "raw.opinioes":     "opinioes.csv",
    "raw.comentarios":  "comentarios.csv",
    "raw.clientes":     "clientes.csv",
    "raw.categorias":   "categorias.csv",
    "raw.restaurantes": "restaurantes.csv",
}
# ==============================
```

```python
from sqlalchemy import create_engine

engine = create_engine(
    MYSQL_DSN,                 # ex: "mysql+pymysql://user:pwd@127.0.0.1:3306/raw"
    pool_pre_ping=True,        # detecta conexão morta antes de usar
    pool_recycle=1800,         # recicla após 30 min (evita timeouts)
    future=True
)


def read_csv_safe(path: str) -> pd.DataFrame:
    # 1) tenta UTF-8
    try:
        return pd.read_csv(
            path,
            sep=None, engine="python",
            on_bad_lines="skip",
            encoding="utf-8",
        )
    except UnicodeDecodeError:
        # 2) tenta Latin-1
        return pd.read_csv(
            path,
            sep=None, engine="python",
            on_bad_lines="skip",
            encoding="latin1",
        )

def load_table(schema_table: str, df: pd.DataFrame):
    schema, table = schema_table.split(".")
    # IMPORTANTÍSSIMO: não transformar tipos/datas aqui
    # para preservar "exatos campos" como estão no CSV.
    # O pandas/SQLAlchemy criará a tabela com esses nomes.
    df.to_sql(
        name=table,
        con=engine,
        schema=schema,
        if_exists="replace",   # cria a tabela do zero com as colunas do DF
        index=False,
        chunksize=500,
        method="multi",
    )

def main():
    # garante que o schema exista (idempotente)
    with engine.begin() as conn:
        conn.exec_driver_sql("CREATE SCHEMA IF NOT EXISTS raw")

    for schema_table, path in CSV_MAP.items():
        p = Path(path)
        if not p.exists():
            print(f"⚠ Arquivo não encontrado: {path} — pulando")
            continue
        print(f"→ Lendo {path} (detectando separador, ignorando linhas ruins)...")
        df = read_csv_safe(path)

        # Se o CSV vier completamente vazio após descartar linhas ruins:
        if df.shape[1] == 0:
            print(f"⚠ {path}: nenhum campo detectado. Verifique o cabeçalho.")
            continue

        print(f"→ Criando/carregando tabela {schema_table} com {len(df)} linhas e {len(df.columns)} colunas…")
        load_table(schema_table, df)
        print(f"✔ {schema_table} pronta.")

    print("✅ Concluído: tabelas em `raw` criadas com colunas exatamente como nos CSVs.")

if __name__ == "__main__":
    main()
```

```python
# CÉLULA DE RESET DO ENGINE
try:
    engine.dispose()  # derruba conexões “sujas” do pool
    print("OK: engine.dispose() executado.")
except NameError:
    print("Variável 'engine' não existe nesta célula/kernel.")
```
